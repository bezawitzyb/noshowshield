```python
#Set missing value for agent, children and company to 0
df['agent'] = df['agent'].fillna(0)
df['children'] = df['children'].fillna(0)
df['company'] = df['company'].fillna(0)
#Drop rows with missing values in the country column
df = df.dropna(subset=['country'])
df.head()
```

```python
#Build preproc pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
numerical_features
#len(numerical_features)
numerical_features_no_target = numerical_features.drop('is_canceled')
```

```python
#Categorical features
categorical_features = df.select_dtypes(include=['object']).columns
categorical_features
#len(categorical_features)
```

```python
#Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
#Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
#Combine preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features_no_target),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
preprocessor
```

```python
df.head()
```

```python
#Create X and y (not transformed yet)
X = df.drop('is_canceled', axis=1)
y = df['is_canceled']
X.head()
```

```python
#Preprocess the data of X
X = preprocessor.fit_transform(X)
```

```python
X.shape
type(X)
```

```python
# Convert X to a DataFrame
X = pd.DataFrame(X.toarray())
X.head()
#Restore column names
#Get the column names from the preprocessor
numerical_columns = numerical_features_no_target
categorical_columns = preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features)
all_columns = np.concatenate([numerical_columns, categorical_columns])
X.columns = all_columns
X.head()
```

```python
import seaborn as sns

# Heatmap of pairwise correlations
correlation_matrix = df.select_dtypes('number').corr()
column_names = correlation_matrix.columns
sns.heatmap(correlation_matrix, xticklabels=column_names, yticklabels=column_names,cmap= "bwr");
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Model to be cross-validated
log_model = LogisticRegression()

#fit the model to the data
log_model.fit(X, y)

# Cross Validation
np.mean(cross_val_score(log_model, X, y , cv=5))
```
