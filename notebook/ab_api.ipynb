```python
from datetime import datetime
from pathlib import Path

import pandas as pd

from eda_package.data import clean_data, load_raw_data, split_X_y
from eda_package.features import engineer_features_v2
from eda_package.main import load_and_preprocess
from eda_package.model import BookingPredictor
from eda_package.preprocessor import transform_preprocessor
from eda_package.registry import COUNTRY_LIMIT, WORKING_MODEL_FILE_NAME
from eda_package.preprocessor import group_countries


def cancellation_probability_series(start_date, end_date):
    """Return cancellation probability for each booking between two dates.

    Parameters
    ----------
    start_date, end_date : str or datetime-like
        Inclusive date range for `arrival_date`.

    Returns
    -------
    pandas.Series
        Probability of cancellation for each row in the filtered data.
    """

    # Build the same preprocessing pipeline used during model training
    X_train_processed, X_test_processed, y_train, y_test, preprocessor = load_and_preprocess()

    df = load_raw_data()
    df = clean_data(df)
    df = group_countries(df, COUNTRY_LIMIT)
    df = engineer_features_v2(df)

    df = df.copy()
    df["arrival_date"] = pd.to_datetime(
        df["arrival_date_year"].astype(str)
        + " "
        + df["arrival_date_month"]
        + " "
        + df["arrival_date_day_of_month"].astype(str),
        format="%Y %B %d",
    )

    start_dt = pd.to_datetime(start_date)
    end_dt = pd.to_datetime(end_date)

    mask = (df["arrival_date"] >= start_dt) & (df["arrival_date"] <= end_dt)
    df = df.loc[mask].copy()

    if df.empty:
        return pd.Series([], dtype=float)

    X, _ = split_X_y(df)
    X_processed = transform_preprocessor(X, preprocessor)

    # Load trained model if exists, otherwise train and save
    try:
        model = BookingPredictor()
    except Exception:
        model = BookingPredictor(X_train_processed=X_train_processed, y_train=y_train)
        model.save_model()

    probs = model.predict_proba(X_processed)[:, 1]
    return pd.Series(probs, index=X.index, name="cancellation_probability")
```

```python
probs = cancellation_probability_series("2017-01-01", "2017-01-01")
probs.head(3)
```

```python
print('Bookings in time period: ', len(probs), '\nAvg. cancellation proba:', probs.mean(), '\nNumber of expected cancellations:', round(len(probs) * probs.mean(),2))
```
